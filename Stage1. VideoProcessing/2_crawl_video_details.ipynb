```python
#crawl thông tin snippet,statistics,contentDetails, videoID từ df['videoId]
def get_video_details(df, start, end, api_key):
    """
    Lấy thông tin chi tiết các video từ DataFrame và lưu vào file JSON.

    Parameters:
    - df: pandas DataFrame có cột 'videoId'
    - start: chỉ số bắt đầu (int)
    - end: chỉ số kết thúc (int)
    - api_key: khóa API YouTube Data API v3 (str)
    """
    video_ids = df['videoId'].iloc[start:end].tolist()

    all_results = []

    for i in range(0, len(video_ids), 50):  # YouTube API giới hạn tối đa 50 ID mỗi request
        batch_ids = video_ids[i:i+50]
        params = {
            "part": "snippet,statistics,contentDetails",
            "id": ",".join(batch_ids),
            "key": api_key
        }
        response = requests.get("https://www.googleapis.com/youtube/v3/videos", params=params)
        
        if response.status_code == 200:
            results = response.json()
            all_results.extend(results.get('items', []))
        else:
            print(f"❌ Lỗi với batch {i}-{i+50}: {response.status_code}")
            print(response.text)

    # Lưu vào file JSON
    output_path = f"../data/raw/detailed_info/video_info_{end}.json"
    with open(output_path, "w", encoding="utf-8") as f:
        json.dump(all_results, f, indent=2, ensure_ascii=False)
    
    print(f"✅ Đã lưu {len(all_results)} video vào {output_path}")
```
