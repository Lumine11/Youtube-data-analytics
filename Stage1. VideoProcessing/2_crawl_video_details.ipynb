```python
#crawl th√¥ng tin snippet,statistics,contentDetails, videoID t·ª´ df['videoId]
def get_video_details(df, start, end, api_key):
    """
    L·∫•y th√¥ng tin chi ti·∫øt c√°c video t·ª´ DataFrame v√† l∆∞u v√†o file JSON.

    Parameters:
    - df: pandas DataFrame c√≥ c·ªôt 'videoId'
    - start: ch·ªâ s·ªë b·∫Øt ƒë·∫ßu (int)
    - end: ch·ªâ s·ªë k·∫øt th√∫c (int)
    - api_key: kh√≥a API YouTube Data API v3 (str)
    """
    video_ids = df['videoId'].iloc[start:end].tolist()

    all_results = []

    for i in range(0, len(video_ids), 50):  # YouTube API gi·ªõi h·∫°n t·ªëi ƒëa 50 ID m·ªói request
        batch_ids = video_ids[i:i+50]
        params = {
            "part": "snippet,statistics,contentDetails",
            "id": ",".join(batch_ids),
            "key": api_key
        }
        response = requests.get("https://www.googleapis.com/youtube/v3/videos", params=params)
        
        if response.status_code == 200:
            results = response.json()
            all_results.extend(results.get('items', []))
        else:
            print(f"‚ùå L·ªói v·ªõi batch {i}-{i+50}: {response.status_code}")
            print(response.text)

    # L∆∞u v√†o file JSON
    output_path = f"../data/raw/detailed_info/video_info_{end}.json"
    with open(output_path, "w", encoding="utf-8") as f:
        json.dump(all_results, f, indent=2, ensure_ascii=False)
    
    print(f"‚úÖ ƒê√£ l∆∞u {len(all_results)} video v√†o {output_path}")
```
